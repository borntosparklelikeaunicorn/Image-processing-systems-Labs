```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
#1.Считать цветное rgb изображение. Преобразовать в градации серого.

# Загрузка изображения
image = cv2.imread('test.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Функция для отображения изображений
def show_images(image, image_2, title=None, title_2=None):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    ax1.imshow(image, cmap='gray')
    ax1.set_title(title)
    ax1.axis('off')

    ax2.imshow(image_2, cmap='gray')
    ax2.set_title(title_2)
    ax2.axis('off')

def rgb_to_gray(image):
    gray_image = np.dot(image[...,:3], [0.2989, 0.5870, 0.1140])
    return gray_image.astype(np.uint8)  # Приведение к целочисленному типу

gray_image = rgb_to_gray(image)
show_images(image, gray_image, title='Оригинальное изображение', title_2='Изображение в градации серого')
```

```python
#2. Сделать выделение контуров методом простого градиента.
#В качестве значения модуля градиента использовать указанный в вариантах метод.

#Модуль градиента аппроксимируется максимумом среди модулей производных

#функция свертки
def convolution(mat1, kernel):
    filtered_image = np.zeros_like(mat1, dtype=np.float32)
    kernel = np.flipud(np.fliplr(kernel))
    pad_y = kernel.shape[0] // 2
    pad_x = kernel.shape[1] // 2
    padded_image = np.pad(mat1, ((pad_y, pad_y), (pad_x, pad_x)), mode='edge')
    for i in range(filtered_image.shape[0]):
        for j in range(filtered_image.shape[1]):
            region = padded_image[i:i + kernel.shape[0], j:j + kernel.shape[1]]
            filtered_image[i, j] = np.sum(region * kernel)
    return filtered_image

#модуль градиента через максимум производных
def gradient_simple_max(gray):
    kernel_x = np.array([[-1, 1]])
    kernel_y = np.array([[-1], [1]])
    gx = convolution(gray, kernel_x)
    gy = convolution(gray, kernel_y)
    grad = np.maximum(np.abs(gx), np.abs(gy))

    binary_grad = np.zeros_like(grad, dtype=np.uint8)
    binary_grad[grad > 25] = 255
    return binary_grad

gradient_image = gradient_simple_max(gray_image)
show_images(gray_image, gradient_image, title='Изображение в градации серого', title_2='Выделение контуров методом простого градиента')
```

```python
#3.Сделать выделение контуров методом по вариантам
#оператор лапласа

def laplace_edge(image):
    #Ядро оператора Лапласа
    laplacian_kernel = np.array([[0,  1, 0],
                                 [1, -4, 1],
                                 [0,  1, 0]])

    laplacian_result = convolution(image, laplacian_kernel)
    laplacian_abs = np.abs(laplacian_result)  #берем модуль значения

    #преобазуем в бинарное изображение с порогом 13
    binary_image = cv2.threshold(laplacian_abs.astype(np.uint8), 13, 255, 0)[1]

    return binary_image
lap_binary = laplace_edge(gray_image)
show_images(gray_image, lap_binary, title='Изображение в градации серого', title_2='Выделение контуров оператором Лапласа')
```

```python
#4.Сделать выделение контуров методом с согласованием. Тип функции аппроксимации и размер окна указан по вариантам.
#Аппроксимация поверхностью 1-го порядка, окно 3x3

#используем "ПреВитта"
def contour_agree_method(image:  np.ndarray , kernel_A = np.array([[-1,-1,-1],[0,0,0],[1,1,1]]),kernel_B = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])):
    finalImage = np.zeros_like(image)
    A = convolution(image, (1/6)*kernel_A)
    B = convolution(image, (1/6)*kernel_B)
    grad = np.sqrt(A ** 2 + B ** 2)
    finalImage[grad > 9] = 255
    return finalImage

agree_met_image = contour_agree_method(gray_image)
show_images(gray_image, agree_met_image, title='Изображение в градации серого', title_2='Контуры по согласованию (аппроксимация 1-го порядка)')
```

```python
#5.Сделать выделение контуров методом Canny (Функция OpenCV).

v = np.median(gray_image) #медианное значение яркости изображения
sigma1 = 0.1 #для более чувствительного порога
sigma2 = 0.9 #для менее чувствительного порога

edge1 = cv2.Canny(gray_image, int(max(0, (1.0 - sigma1) * v)), int(min(255, (1.0 + sigma1) * v)))
edge2 = cv2.Canny(gray_image,  int(max(0, (1.0 - sigma2) * v)), int(min(255, (1.0 + sigma2) * v)))

plt.figure(figsize=(14, 7))
plt.subplot(1, 3, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Исходное изображение')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(edge1, cmap='gray')
plt.title('Метод Canny1')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(edge2, cmap='gray')
plt.title('Метод Canny2')
plt.axis('off')
```
